```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("2016_raw_08062019.xlsx")
```

```python
df.head()
```

```python
df.iloc[:,6:].head()
```

```python
# Algorithm terminal layout 2 for saving the latitude and longitude value to be called when making dataframe
dict_loc_origin = {df['O_City'][x]:(df['O_City_Latitude'][x], df['O_City_Longitude'][x]) for x in range(len(df))}
dict_loc_destination = {df['D_City'][x]:(df['D_City_Latitude'][x], df['D_City_Longitude'][x]) for x in range(len(df))}
```

```python
# Sum untuk quantify jumlah departure per week dan transport time days
# Count untuk menghitung jumlah service
df_group = df.groupby(by=["O_City","D_City","Transportation mode","Carrier"]).count()
```

```python
df_group.head()
```

```python
# Make another dataframe to group only 3 index
df_group_3index = df.groupby(by=["O_City","D_City","Transportation mode"]).count()
```

```python
three_level_index = list(df_group_3index.index)
```

```python
three_level_index[:5]
```

```python
# testing
df_group.loc[three_level_index[0],:]
```

```python
dict_carrier_number = dict()
for i, j in enumerate(three_level_index):
    dict_carrier_number[j] = len(df_group.loc[three_level_index[i],:])
```

```python
dict_carrier_number[('Rotterdam','Antwerp','Sea')]
```

```python
# Make new dataframe to contain the result
df_carrier_number = pd.DataFrame(columns=['O_City','D_City','Modality','No_carrier','departure_per_week','avg_travel_time','link_weight','o_city_lat','o_city_lon','d_city_lat','d_city_lon'], index=range(len(three_level_index)))
```

```python
df_carrier_number.head()
```

```python
for i, j in enumerate(three_level_index):
    df_carrier_number['O_City'].iloc[i] = three_level_index[i][0]
    df_carrier_number['D_City'].iloc[i] = three_level_index[i][1]
    df_carrier_number['Modality'].iloc[i] = three_level_index[i][2]
    df_carrier_number['No_carrier'].iloc[i] = dict_carrier_number[j]
```

```python
df_carrier_number[df_carrier_number['O_City'] == 'Rotterdam'].head()
```

```python
# Fill the coordinate
for i in range(len(df_carrier_number)):
    df_carrier_number['o_city_lat'].iloc[i] = dict_loc_origin[df_carrier_number['O_City'].iloc[i]][0]
    df_carrier_number['o_city_lon'].iloc[i] = dict_loc_origin[df_carrier_number['O_City'].iloc[i]][1]
    df_carrier_number['d_city_lat'].iloc[i] = dict_loc_destination[df_carrier_number['D_City'].iloc[i]][0]
    df_carrier_number['d_city_lon'].iloc[i] = dict_loc_destination[df_carrier_number['D_City'].iloc[i]][1]
```

```python
# Combine information about number of frequency to the dataframe
```

```python
df_freq = df.groupby(by=["O_City","D_City","Transportation mode"]).sum()
```

```python
df_time = df.groupby(by=["O_City","D_City","Transportation mode"]).mean()
```

```python
df_freq.loc[('Rotterdam','Antwerp','Sea'),:]['Number of departures per week']
```

```python
df_time.loc[('Rotterdam','Antwerp','Sea'),:]['Transport time (days)']
```

```python
for i, j in enumerate(three_level_index):
    df_carrier_number['departure_per_week'].iloc[i] = df_freq.loc[j]["Number of departures per week"]
    df_carrier_number['avg_travel_time'].iloc[i] = df_time.loc[j]["Transport time (days)"]
```

```python
df_carrier_number[df_carrier_number['O_City'] == 'Rotterdam'].head()
```
